```python
import pandas as pd
df = pd.read_excel('2 tables test task.xlsx')

df['month'] = df['month'].astype('str')
df['month.1'] = df['month.1'].astype('str')

df2 = (df[['month.1', 'platform.1', 'c']]
        .rename(columns = {'month.1': 'month', 'platform.1': 'platform'})
        .set_index(['month', 'platform'])
        .sort_index())

df = (df.set_index(['month', 'platform'])
        .sort_index())[['a', 'b']]

merged = (df.merge(df2, on = ['month', 'platform'], how = 'outer')
            .dropna(how = 'all')
            .reset_index())
rearranged_cols = ['month', 'platform', 'a', 'c', 'b']
merged = merged[rearranged_cols]

merged.to_excel('output.xlsx', index = False)
merged
```
